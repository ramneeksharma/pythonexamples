```python
img_nz_bgr = cv2.imread("new_zealand_lake.jpg", cv2.IMREAD_COLOR)
b, g, r = cv2.split(img_nz_bgr)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap='gray');plt.title("Red Channel");
plt.subplot(142);plt.imshow(b,cmap='gray');plt.title("Blue Channel");
plt.subplot(143);plt.imshow(g,cmap='gray');plt.title("Green Channel")

img_merged = cv2.merge((b,g,r))
plt.subplot(144);plt.imshow(img_merged[:,:,::-1]);plt.title("Merged Channel")
```

```python
img_hsv = cv2.cvtColor(img_nz_bgr, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(img_hsv)
plt.figure(figsize=[20,5])
plt.subplot(147);plt.imshow(h, cmap='gray');plt.title("H Channel"); 
```
